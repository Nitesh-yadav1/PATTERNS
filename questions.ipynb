```python
1.Print Number Pyramid
Send Feedback
Print the following pattern for a given n.
For eg. N = 6
123456
  23456
    3456
      456
        56
          6
        56
      456
    3456
  23456
123456

n=int(input())
n1=n
n2=n-1
k=n2
for i in range(1,n1+1,1):
    for s in range(i-1):
        print(" ",end="")
    for j in range(n1-i+1):
        print(i,end="")
        i=i+1
    print()
    
for i in range(1,n2+1,1):
    for s in range(n2-i):
        print(" ",end="")
    for j in range(i+1):
        print(k,end="")
        k=k+1
    k=n2-i
    print()
```

```python
2.Binary Pattern
Send Feedback
Print the following pattern for the given number of rows.
Pattern for N = 4
1111
000
11
0

n=int(input())
for i in range(1,n+1):
    for j in range(1,n-i+2):
        if(i%2==1):
            print("1",end="")
        else:
            print("0",end='')
    print()
```

```python
3.Zeros and Stars Pattern
Send Feedback
Print the following pattern
Pattern for N = 4
*000*000*
0*00*00*0
00*0*0*00
000***000

n=int(input())
i=1
j=1
while i<=n:
    j=1
    while j<=n:
        if i==j:
            print("*",end='',flush=True)
        else:
            print("0",end='',flush=True)
        j=j+1
    j=j-1
    print("*",end='',flush=True)
    while j>=1:
        if i==j:
            print("*",end='',flush=True)
        else:
            print("0",end='',flush=True)
        j=j-1
    print();
    i=i+1
```

```python
4.Alpha Pattern
Send Feedback
Print the following pattern for the given N number of rows.
Pattern for N = 3
 A
 BB
 CCC
n=int(input())
i=1
while i<=n:
    j=1
    start_char=chr(ord('A')+i-1)
    while j<=i:
        print(start_char,end='')
        j=j+1
    print()
    i=i+1
```

```python
10.Write a program to determine if given number is palindrome or not. Print true if it is palindrome, false otherwise.

n=int(input())
temp=n
rev=0

while temp!=0:
    rev=(rev*10) + (temp%10)
    temp=temp//10
if (n==rev):
    print('true')
else:
    print('false')
```
